```python
# --- CELL 1: IMPORT THƯ VIỆN CHO RANDOM FOREST VÀ CÁC TÁC VỤ LIÊN QUAN ---

# Thư viện cơ bản cho xử lý dữ liệu và tính toán
import pandas as pd
import numpy as np

# Thư viện cho trực quan hóa
import matplotlib.pyplot as plt
import seaborn as sns # Hữu ích cho một số biểu đồ EDA và phân tích phần dư

# Thư viện Scikit-learn
from sklearn.ensemble import RandomForestRegressor # Mô hình Random Forest
from sklearn.model_selection import train_test_split # Dùng để chia train/test (cần cẩn thận với chuỗi thời gian)
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score # Các độ đo đánh giá
import statsmodels.api as sm
# (Tùy chọn) Statsmodels cho phân tích phần dư nâng cao (nếu cần)
# import statsmodels.api as sm # Cho Q-Q plot
# from statsmodels.graphics.tsaplots import plot_acf # Cho ACF của phần dư
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler # Để chuẩn hóa dữ liệu

# Cài đặt cho Matplotlib để hiển thị biểu đồ ngay trong notebook
%matplotlib inline

# Cài đặt cho Pandas để hiển thị số float (tùy chọn)
pd.set_option('display.float_format', lambda x: '%.3f' % x)
# pd.set_option('display.max_rows', 100)
# pd.set_option('display.max_columns', 50)

print("--- Cell 1: Import Thư viện cho Random Forest Hoàn Tất ---")
```

```python
# --- PHẦN 1: ĐỌC VÀ TIỀN XỬ LÝ DỮ LIỆU CHUNG ---

print("--- BẮT ĐẦU QUÁ TRÌNH XỬ LÝ DỮ LIỆU ---")

# 1.1. Đọc dữ liệu từ file CSV
file_path = 'World GDP Dataset.csv'
try:
    df_raw_initial = pd.read_csv(file_path)
    df_raw = pd.read_csv(file_path, na_values=['0', 0.0])
except FileNotFoundError:
    print(f"Lỗi: Không tìm thấy file '{file_path}'. Vui lòng kiểm tra lại đường dẫn.")
    raise
except Exception as e:
    print(f"Lỗi khi đọc file CSV: {e}")
    raise
print("Đọc dữ liệu thô thành công.")

print("\n--- THỐNG KÊ DỮ LIỆU THÔ BAN ĐẦU (df_raw_initial - TRƯỚC KHI XỬ LÝ SÂU) ---")
rows_raw_initial, cols_raw_initial = df_raw_initial.shape
elements_raw_initial = df_raw_initial.size
print(f"Kích thước: {rows_raw_initial} hàng × {cols_raw_initial} cột ({elements_raw_initial} phần tử)")
print("Hiển thị tóm tắt dữ liệu thô ban đầu:")
# Pandas sẽ tự động tóm tắt nếu số hàng > pd.options.display.max_rows (thường là 60)
# Nếu muốn luôn thấy kiểu tóm tắt ngay cả với df nhỏ hơn, bạn cần thủ thuật hơn
# hoặc chấp nhận hiển thị toàn bộ nếu nó nhỏ.
# Với df_raw_initial có 230 hàng, nó sẽ tự tóm tắt.
if 'df_raw_initial' in locals():
    with pd.option_context('display.show_dimensions', True): # Hiển thị kích thước ở cuối
        display(df_raw_initial)
else:
    print("df_raw_initial chưa được tạo.")


# 1.2. Làm sạch cơ bản (áp dụng cho df_raw)
df_raw.dropna(how='all', inplace=True)
rows_before_copyright_removal = len(df_raw)
df_raw = df_raw[~df_raw.iloc[:, 0].astype(str).str.contains("©IMF", na=False)]
rows_after_copyright_removal = len(df_raw)
print(f"\nLàm sạch cơ bản: Loại bỏ {rows_before_copyright_removal - rows_after_copyright_removal} dòng copyright/footer.")

country_col_original_name = df_raw.columns[0]
df_raw.rename(columns={country_col_original_name: 'Country Name'}, inplace=True)
print("Đổi tên cột quốc gia hoàn tất.")

print("\n--- THỐNG KÊ DỮ LIỆU SAU LÀM SẠCH CƠ BẢN (df_raw - TRƯỚC KHI MELT) ---")
rows_raw_cleaned, cols_raw_cleaned = df_raw.shape
elements_raw_cleaned = df_raw.size
print(f"Kích thước: {rows_raw_cleaned} hàng × {cols_raw_cleaned} cột ({elements_raw_cleaned} phần tử)")
print("Hiển thị tóm tắt dữ liệu sau làm sạch cơ bản:")
if 'df_raw' in locals():
    with pd.option_context('display.show_dimensions', True):
        display(df_raw) # df_raw có 196 hàng, cũng sẽ tự tóm tắt
else:
    print("df_raw (đã làm sạch) chưa được tạo.")


# 1.3. Chuyển từ định dạng wide sang long
try:
    df_long = pd.melt(df_raw,
                      id_vars=['Country Name'],
                      var_name='Year',
                      value_name='GDP')
except KeyError as e:
    print(f"Lỗi khi chuyển đổi wide sang long: Cột 'Country Name' hoặc các cột năm không đúng. {e}")
    raise
print("\nChuyển đổi dữ liệu từ wide sang long thành công.")
print("--- THỐNG KÊ DỮ LIỆU SAU KHI MELT (df_long - TRƯỚC KHI CHUYỂN KIỂU VÀ DROPNA) ---")
rows_after_melt, cols_after_melt = df_long.shape
elements_after_melt = df_long.size
print(f"Kích thước: {rows_after_melt} hàng × {cols_after_melt} cột ({elements_after_melt} phần tử)")
print("Hiển thị tóm tắt dữ liệu sau khi melt:")
if 'df_long' in locals():
    with pd.option_context('display.show_dimensions', True):
        # df_long có 8624 hàng, chắc chắn sẽ được Pandas tóm tắt
        display(df_long)
else:
    print("df_long (sau melt) chưa được tạo.")


# 1.4. Chuyển đổi kiểu dữ liệu
df_long['Year'] = pd.to_numeric(df_long['Year'], errors='coerce')
if df_long['GDP'].dtype == 'object':
    df_long['GDP'] = df_long['GDP'].astype(str).str.replace(',', '', regex=False)
df_long['GDP'] = pd.to_numeric(df_long['GDP'], errors='coerce')
print("\nChuyển đổi kiểu dữ liệu cho 'Year' và 'GDP' hoàn tất.")

# 1.5. Xử lý giá trị thiếu (NaN)
rows_before_final_dropna = len(df_long)
df_long.dropna(subset=['Year', 'GDP'], inplace=True)
rows_after_final_dropna = len(df_long)
print(f"Xử lý giá trị thiếu cuối cùng: Số dòng trước khi loại bỏ NaN ở 'Year'/'GDP' (sau chuyển kiểu): {rows_before_final_dropna}")
print(f"Số dòng sau khi loại bỏ NaN cuối cùng: {rows_after_final_dropna}")
print(f"Số dòng đã bị loại bỏ do NaN ở 'Year'/'GDP' hoặc do '0' ban đầu: {rows_before_final_dropna - rows_after_final_dropna}")


# 1.6. Sắp xếp dữ liệu
df_long.sort_values(by=['Country Name', 'Year'], inplace=True)
df_long.reset_index(drop=True, inplace=True)
print("\nSắp xếp dữ liệu theo Quốc gia và Năm hoàn tất.")

print("\n\n--- TIỀN XỬ LÝ DỮ LIỆU HOÀN TẤT (df_long CUỐI CÙNG) ---")
print("--- THỐNG KÊ CHI TIẾT VỀ df_long SAU TẤT CẢ CÁC BƯỚC XỬ LÝ ---")

num_rows_final, num_cols_final = df_long.shape
elements_final = df_long.size
print(f"Kích thước: {num_rows_final} hàng × {num_cols_final} cột ({elements_final} phần tử)")
print("Hiển thị tóm tắt df_long cuối cùng:")
if 'df_long' in locals():
    with pd.option_context('display.show_dimensions', True):
        # df_long có 7852 hàng, chắc chắn sẽ được Pandas tóm tắt
        display(df_long)
else:
    print("df_long (cuối cùng) chưa được tạo.")

# Các thống kê khác giữ nguyên
print(f"\nTên các cột cuối cùng: {df_long.columns.tolist()}")
print("\nKiểu dữ liệu của các cột cuối cùng:")
print(df_long.dtypes)
print("\nSố lượng giá trị duy nhất cho mỗi cột cuối cùng:")
for col in df_long.columns:
    print(f"  - {col}: {df_long[col].nunique()}")
print("\nKiểm tra giá trị thiếu (NaN) trong df_long cuối cùng:")
print(df_long.isnull().sum())
print("\nThống kê mô tả cho các cột số cuối cùng:")
if 'df_long' in locals(): display(df_long.describe())
min_year = df_long['Year'].min() if 'df_long' in locals() and not df_long.empty else 'N/A'
max_year = df_long['Year'].max() if 'df_long' in locals() and not df_long.empty else 'N/A'
print(f"\nPhạm vi năm trong dữ liệu cuối cùng: Từ {min_year} đến {max_year}")
num_unique_countries = df_long['Country Name'].nunique() if 'df_long' in locals() and not df_long.empty else 'N/A'
print(f"Tổng số quốc gia có dữ liệu cuối cùng: {num_unique_countries}")
```

```python
# === THAM SỐ CHUNG CHO MÔ HÌNH SVR NÀY ===
MODEL_ID_M6 = "M6_SVR_LagsYear" # Đổi tên để phản ánh đặc trưng mới
COUNTRY_M6 = 'Vietnam' # <<--- THAY ĐỔI QUỐC GIA Ở ĐÂY
TEST_SIZE_RATIO_M6 = 0.25
N_FUTURE_YEARS_M6 = 5
NUM_LAGS_M6 = 5
# ==========================================

print(f"\n\n--- BẮT ĐẦU {MODEL_ID_M6} CHO {COUNTRY_M6.upper()} ---")
print(f"Sử dụng {NUM_LAGS_M6} lags, Year_Feat, và Year_Sq_Feat làm đặc trưng.")

# 1. Lọc dữ liệu cho quốc gia được chọn
if 'df_long' not in locals() or df_long.empty:
    print(f"Lỗi: df_long chưa được tạo hoặc rỗng. Vui lòng chạy Cell 2 trước.")
    X_m6, y_m6, country_data_processed_m6, feature_columns_m6, country_data_svr_m6 = None, None, None, None, None
else:
    country_data_svr_m6 = df_long[df_long['Country Name'] == COUNTRY_M6].copy()
    country_data_svr_m6.sort_values('Year', inplace=True)
    country_data_svr_m6.reset_index(drop=True, inplace=True)

    if country_data_svr_m6.empty or len(country_data_svr_m6) < NUM_LAGS_M6 + 5:
        print(f"Không đủ dữ liệu cho '{COUNTRY_M6}' ({len(country_data_svr_m6)} điểm) để tạo {NUM_LAGS_M6} lags và xây dựng {MODEL_ID_M6}.")
        X_m6, y_m6, country_data_processed_m6, feature_columns_m6 = None, None, None, None
    else:
        print(f"Đã lọc dữ liệu cho {COUNTRY_M6}. Số dòng ban đầu: {len(country_data_svr_m6)}")
        
        plt.figure(figsize=(10, 5))
        plt.plot(country_data_svr_m6['Year'], country_data_svr_m6['GDP'], marker='.', linestyle='-', label=f'GDP Thực tế {COUNTRY_M6}')
        plt.title(f'GDP Thực tế của {COUNTRY_M6} (cho {MODEL_ID_M6})'); plt.xlabel('Năm'); plt.ylabel('GDP (Tỷ USD)'); plt.grid(True); plt.legend(); plt.show()

        # 2. Tạo các đặc trưng
        feature_columns_m6 = []
        # Tạo Lags
        for i in range(1, NUM_LAGS_M6 + 1):
            col_name = f'GDP_lag{i}'
            country_data_svr_m6[col_name] = country_data_svr_m6['GDP'].shift(i)
            feature_columns_m6.append(col_name)
        
        # Tạo đặc trưng Year
        country_data_svr_m6['Year_Feat'] = country_data_svr_m6['Year']
        feature_columns_m6.append('Year_Feat')
        
        # === THÊM ĐẶC TRƯNG YEAR^2 ===
        country_data_svr_m6['Year_Sq_Feat'] = country_data_svr_m6['Year'] ** 2
        feature_columns_m6.append('Year_Sq_Feat')
        # ==============================
        
        print(f"\nĐã tạo các đặc trưng: {feature_columns_m6}.")
        
        # 3. Xử lý giá trị NaN
        country_data_processed_m6 = country_data_svr_m6.dropna().copy()
        country_data_processed_m6.reset_index(drop=True, inplace=True)

        if len(country_data_processed_m6) < NUM_LAGS_M6 + 5: # Kiểm tra lại sau dropna
            print(f"Không đủ dữ liệu cho '{COUNTRY_M6}' ({len(country_data_processed_m6)} dòng) sau khi tạo features và dropna.")
            X_m6, y_m6 = None, None
        else:
            print(f"\nSố dòng dữ liệu của {COUNTRY_M6} sau khi tạo features và dropna: {len(country_data_processed_m6)}")
            
            # 4. Chuẩn bị X (features) và y (target)
            y_m6 = country_data_processed_m6['GDP']
            X_m6 = country_data_processed_m6[feature_columns_m6]

            print(f"\n--- Đặc trưng X cho {MODEL_ID_M6} (5 dòng đầu) ---")
            display(X_m6.head())
            print(f"Các cột đặc trưng: {X_m6.columns.tolist()}")
            print(f"\n--- Mục tiêu y cho {MODEL_ID_M6} (5 dòng đầu) ---")
            display(y_m6.head().to_frame().T)
```

```python
# --- CELL SVR2.M6: {MODEL_ID_M6} - CHIA TRAIN/TEST, CHUẨN HÓA VÀ HUẤN LUYỆN ---

model_svr_m6 = None 
X_train_m6, X_test_m6, y_train_m6, y_test_m6 = None, None, None, None
years_train_m6, years_test_m6 = None, None
scaler_X_m6 = None # Scaler cho X
scaler_y_m6 = None # Scaler cho y (tùy chọn, nhưng hữu ích nếu y có dải rộng)

if 'X_m6' in locals() and X_m6 is not None and \
   'y_m6' in locals() and y_m6 is not None and not X_m6.empty:
    
    # 1. Phân chia dữ liệu Train/Test
    split_index_m6 = int(len(X_m6) * (1 - TEST_SIZE_RATIO_M6))
    X_train_m6_raw, X_test_m6_raw = X_m6.iloc[:split_index_m6], X_m6.iloc[split_index_m6:]
    y_train_m6_raw, y_test_m6_raw = y_m6.iloc[:split_index_m6], y_m6.iloc[split_index_m6:]

    years_full_processed_m6 = country_data_processed_m6['Year']
    years_train_m6 = years_full_processed_m6.iloc[:split_index_m6]
    years_test_m6 = years_full_processed_m6.iloc[split_index_m6:]

    if X_train_m6_raw.empty or X_test_m6_raw.empty or len(X_test_m6_raw) < 1:
        print(f"Không đủ dữ liệu để chia train/test cho {MODEL_ID_M6}.")
    else:
        print(f"{MODEL_ID_M6} - Chia dữ liệu: Train {len(X_train_m6_raw)} mẫu, Test {len(X_test_m6_raw)} mẫu.")

        # 2. Chuẩn hóa Dữ liệu (Scaling) - RẤT QUAN TRỌNG CHO SVR
        scaler_X_m6 = StandardScaler()
        X_train_m6 = scaler_X_m6.fit_transform(X_train_m6_raw)
        X_test_m6 = scaler_X_m6.transform(X_test_m6_raw)
        print("Đặc trưng X_train và X_test đã được chuẩn hóa (StandardScaler).")

        # Tùy chọn: Chuẩn hóa y (biến mục tiêu)
        # Điều này có thể giúp SVR hội tụ tốt hơn nếu y có dải giá trị lớn
        # Nhớ rằng bạn sẽ cần inverse_transform y_pred sau này
        scaler_y_m6 = StandardScaler()
        y_train_m6 = scaler_y_m6.fit_transform(y_train_m6_raw.values.reshape(-1, 1)).flatten()
        # Không fit scaler_y trên y_test_m6_raw, chỉ transform
        # y_test_m6 = scaler_y_m6.transform(y_test_m6_raw.values.reshape(-1, 1)).flatten()
        # Tuy nhiên, để đánh giá, chúng ta sẽ dùng y_test_m6_raw (thang đo gốc)
        # và biến đổi ngược y_pred về thang đo gốc.
        # Vì vậy, y_test_m6 sẽ là y_test_m6_raw để tiện so sánh.
        y_test_m6 = y_test_m6_raw # Giữ y_test ở thang đo gốc để đánh giá

        print("Biến mục tiêu y_train đã được chuẩn hóa. y_test giữ thang đo gốc để đánh giá.")


        # 3. Huấn luyện mô hình SVR
        # Các siêu tham số cơ bản, có thể cần tinh chỉnh (ví dụ: dùng GridSearchCV)
        # Kernel 'rbf' (Radial Basis Function) thường linh hoạt
        # C và gamma là các siêu tham số quan trọng cho kernel 'rbf'
        # Epsilon xác định vùng không phạt lỗi
        #model_svr_m6 = SVR(kernel='rbf', C=5.0, epsilon=0.1, gamma=0.1) 
        # Các kernel khác: 'linear', 'poly'
        # Với dữ liệu ít, kernel 'linear' có thể ổn định hơn:
        model_svr_m6 = SVR(kernel='linear', C=1.0, epsilon=0.1)
        
        print(f"\n--- {MODEL_ID_M6}: Bắt đầu huấn luyện SVR (kernel='{model_svr_m6.kernel}') ---")
        model_svr_m6.fit(X_train_m6, y_train_m6) # Huấn luyện trên X, y đã chuẩn hóa
        print(f"Huấn luyện {MODEL_ID_M6} hoàn tất.")
else:
    print(f"Biến X_m6 và y_m6 cho {MODEL_ID_M6} chưa được tạo hoặc rỗng.")
```

```python
# --- CELL SVR3.M6: {MODEL_ID_M6} - ĐÁNH GIÁ MÔ HÌNH ---

r2_svr_m6, mae_svr_m6, rmse_svr_m6, mape_svr_m6 = np.nan, np.nan, np.nan, np.nan
y_pred_test_orig_m6 = None # Dự đoán trên thang đo gốc

if 'model_svr_m6' in locals() and model_svr_m6 is not None and \
   'X_test_m6' in locals() and X_test_m6 is not None and not X_test_m6_raw.empty and \
   'scaler_y_m6' in locals() and scaler_y_m6 is not None: # Cần scaler_y để biến đổi ngược
    
    # 1. Dự đoán trên tập kiểm tra (giá trị dự đoán đang ở thang đo đã chuẩn hóa của y)
    y_pred_test_scaled_m6 = model_svr_m6.predict(X_test_m6)

    # 2. Biến đổi ngược dự đoán về thang đo gốc
    y_pred_test_orig_m6 = scaler_y_m6.inverse_transform(y_pred_test_scaled_m6.reshape(-1, 1)).flatten()
    
    # y_test_m6 (y_test_m6_raw) đã ở thang đo gốc
    y_true_test_orig_m6 = y_test_m6_raw.values # Lấy numpy array để tính toán

    print(f"\n--- {MODEL_ID_M6}: Đánh giá trên tập Kiểm tra (thang đo gốc) ---")
    r2_svr_m6 = r2_score(y_true_test_orig_m6, y_pred_test_orig_m6)
    mae_svr_m6 = mean_absolute_error(y_true_test_orig_m6, y_pred_test_orig_m6)
    rmse_svr_m6 = np.sqrt(mean_squared_error(y_true_test_orig_m6, y_pred_test_orig_m6))
    
    y_true_m = pd.Series(y_true_test_orig_m6); y_pred_m = pd.Series(y_pred_test_orig_m6) # Chuyển sang Series để dễ lọc
    zero_mask = (np.abs(y_true_m) < 1e-9)
    y_true_f = y_true_m[~zero_mask]; y_pred_f = y_pred_m[~zero_mask]
    if len(y_true_f) > 0: mape_svr_m6 = np.mean(np.abs((y_true_f - y_pred_f) / y_true_f)) * 100
    else: mape_svr_m6 = np.nan
            
    print(f"R-squared (Test): {r2_svr_m6:.4f}")
    print(f"MAE (Test): {mae_svr_m6:.2f}")
    print(f"RMSE (Test): {rmse_svr_m6:.2f}")
    if not np.isnan(mape_svr_m6): print(f"MAPE (Test): {mape_svr_m6:.2f}% (Tính trên {len(y_true_f)} điểm)")
    else: print("MAPE (Test): Không thể tính.")

    # 3. Trực quan hóa kết quả trên tập test
    plt.figure(figsize=(14, 7))
    if 'years_train_m6' in locals() and years_train_m6 is not None and 'y_train_m6_raw' in locals(): # Vẽ train gốc
        plt.plot(years_train_m6, y_train_m6_raw, label='GDP Huấn luyện (Thực tế)', color='royalblue', marker='o', markersize=4)
    if 'years_test_m6' in locals() and years_test_m6 is not None:
        plt.plot(years_test_m6, y_true_test_orig_m6, label='GDP Kiểm tra (Thực tế)', color='forestgreen', marker='o', markersize=4)
        plt.plot(years_test_m6, y_pred_test_orig_m6, label=f'GDP Kiểm tra (Dự đoán {MODEL_ID_M6})', color='purple', marker='x', markersize=6, linestyle='--')
    
    plt.title(f'{MODEL_ID_M6}: Đánh giá cho {COUNTRY_M6}', fontsize=15)
    plt.xlabel('Năm (Year)', fontsize=12); plt.ylabel('GDP (Tỷ Đô la Mỹ)', fontsize=12)
    plt.legend(fontsize=10); plt.grid(True, linestyle=':', alpha=0.7); plt.xticks(rotation=30, ha='right'); plt.tight_layout(); plt.show()
else:
    print(f"Mô hình {MODEL_ID_M6} chưa được huấn luyện hoặc không có dữ liệu/scaler để đánh giá.")
```

```python
# --- CELL SVR4.M6: {MODEL_ID_M6} - PHÂN TÍCH PHẦN DƯ ---

if 'model_svr_m6' in locals() and model_svr_m6 is not None and \
   'y_test_m6_raw' in locals() and y_test_m6_raw is not None and \
   'y_pred_test_orig_m6' in locals() and y_pred_test_orig_m6 is not None:
    
    print(f"\n\n--- {MODEL_ID_M6}: PHÂN TÍCH PHẦN DƯ CHO {COUNTRY_M6.upper()} (TRÊN TẬP TEST, THANG ĐO GỐC) ---")
    
    # Phần dư được tính trên thang đo gốc
    residuals_test_m6 = y_test_m6_raw.values - y_pred_test_orig_m6 # y_test_m6_raw là Series, y_pred_test_orig_m6 là array

    # 1. Biểu đồ Phân tán Phần dư (Residuals vs. Predicted Values - thang đo gốc)
    plt.figure(figsize=(10, 5))
    plt.scatter(y_pred_test_orig_m6, residuals_test_m6, alpha=0.7, label='Tập Kiểm tra', color='green')
    plt.axhline(y=0, color='r', linestyle='--'); plt.xlabel('GDP Dự đoán (Thang đo gốc)'); plt.ylabel('Phần dư (Thang đo gốc)')
    plt.title(f'{MODEL_ID_M6}: Phần dư vs. Dự đoán (Test - Gốc)'); plt.legend(); plt.grid(True); plt.show()

    # 2. Biểu đồ Phân phối của Phần dư (Histogram và Q-Q Plot - trên tập test)
    plt.figure(figsize=(12, 5)); plt.subplot(1, 2, 1)
    sns.histplot(residuals_test_m6, bins=max(5, len(residuals_test_m6)//2), kde=True, edgecolor='black', color='green') # Điều chỉnh bins
    plt.xlabel(f'Phần dư {MODEL_ID_M6} (Test - Gốc)'); plt.ylabel('Tần suất'); 
    plt.title(f'{MODEL_ID_M6}: Phân phối Phần dư (Test - Gốc)'); plt.grid(axis='y');
    plt.subplot(1, 2, 2); sm.qqplot(residuals_test_m6, line='s', ax=plt.gca()); 
    plt.title(f'{MODEL_ID_M6}: Q-Q Plot Phần dư (Test - Gốc)'); plt.grid(True)
    plt.tight_layout(); plt.show()

    # 3. Biểu đồ Phần dư theo Thời gian (trên tập test)
    plt.figure(figsize=(14, 6))
    if 'years_test_m6' in locals() and years_test_m6 is not None and len(years_test_m6) == len(residuals_test_m6):
        plt.plot(years_test_m6, residuals_test_m6, marker='x', linestyle='-', color='green', label='Phần dư Kiểm tra')
        plt.axhline(y=0, color='r', linestyle='--'); plt.xlabel('Năm'); plt.ylabel('Phần dư')
        plt.title(f'{MODEL_ID_M6}: Phần dư theo Thời gian cho {COUNTRY_M6} (Test - Gốc)'); plt.legend(); plt.grid(True); plt.tight_layout(); plt.show()
    else:
        print("Không thể vẽ biểu đồ phần dư theo thời gian do thiếu thông tin năm hoặc độ dài không khớp.")
else:
    print(f"Mô hình {MODEL_ID_M6} chưa có đủ kết quả để phân tích phần dư.")
```
