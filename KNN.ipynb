```python
# === THAM SỐ CHUNG CHO MÔ HÌNH NÀY ===
MODEL_ID_M7D = "M7_KNN_DiffLags"
COUNTRY_M7D = 'Vietnam' # <<--- THAY ĐỔI QUỐC GIA Ở ĐÂY
TEST_SIZE_RATIO_M7D = 0.2
N_FUTURE_YEARS_M7D = 5
NUM_DIFF_LAGS_M7D = 3 # Số lags của chuỗi sai phân GDP_diff1
N_NEIGHBORS_M7D = 5   # Số láng giềng cho KNN
# =====================================

print(f"\n\n--- BẮT ĐẦU {MODEL_ID_M7D} CHO {COUNTRY_M7D.upper()} ---")
print(f"Sử dụng {NUM_DIFF_LAGS_M7D} lags của GDP_diff1. K={N_NEIGHBORS_M7D}. Dữ liệu X sẽ được chuẩn hóa.")

# 1. Lọc dữ liệu và tạo đặc trưng sai phân
X_m7d, y_m7d_diff_target, country_data_processed_m7d, feature_columns_m7d, country_data_knn_m7d = None, None, None, None, None
# Biến để lưu GDP(t-1) cho việc tích hợp ngược
gdp_t_minus_1_for_integration_m7d = None

if 'df_long' not in locals() or df_long.empty:
    print(f"Lỗi: df_long chưa được tạo. Vui lòng chạy Cell 2.")
else:
    country_data_knn_m7d = df_long[df_long['Country Name'] == COUNTRY_M7D].copy()
    country_data_knn_m7d.sort_values('Year', inplace=True)
    country_data_knn_m7d.reset_index(drop=True, inplace=True)

    min_data_len_m7d = NUM_DIFF_LAGS_M7D + 1 + 5 # Cần 1 cho diff, NUM_DIFF_LAGS_M7D cho lags, và ít nhất 5 cho train/test
    if country_data_knn_m7d.empty or len(country_data_knn_m7d) < min_data_len_m7d:
        print(f"Không đủ dữ liệu cho '{COUNTRY_M7D}' ({len(country_data_knn_m7d)} điểm, cần ít nhất ~{min_data_len_m7d}) để xây dựng {MODEL_ID_M7D}.")
    else:
        print(f"Đã lọc dữ liệu cho {COUNTRY_M7D}. Số dòng ban đầu: {len(country_data_knn_m7d)}")
        
        # 1a. Tạo chuỗi sai phân bậc 1
        country_data_knn_m7d['GDP_diff1'] = country_data_knn_m7d['GDP'].diff(1)
        
        # 1b. Tạo các lags của chuỗi sai phân GDP_diff1
        feature_columns_m7d = []
        for i in range(1, NUM_DIFF_LAGS_M7D + 1):
            col_name = f'GDP_diff1_lag{i}'
            country_data_knn_m7d[col_name] = country_data_knn_m7d['GDP_diff1'].shift(i)
            feature_columns_m7d.append(col_name)
        
        print(f"\nĐã tạo các đặc trưng sai phân: {feature_columns_m7d}.")
        
        # 1c. Xử lý giá trị NaN (do diff và shift)
        # Giữ lại cột 'Year' và 'GDP' gốc để tham chiếu và tích hợp ngược
        country_data_processed_m7d = country_data_knn_m7d.dropna(subset=feature_columns_m7d + ['GDP_diff1']).copy()
        country_data_processed_m7d.reset_index(drop=True, inplace=True)

        if len(country_data_processed_m7d) < NUM_DIFF_LAGS_M7D + 5: 
            print(f"Không đủ dữ liệu ({len(country_data_processed_m7d)} dòng) sau khi tạo features và dropna cho {MODEL_ID_M7D}.")
        else:
            print(f"\nSố dòng dữ liệu của {COUNTRY_M7D} sau khi tạo features và dropna: {len(country_data_processed_m7d)}")
            
            # 1d. Chuẩn bị X (lags của GDP_diff1) và y (GDP_diff1 hiện tại)
            y_m7d_diff_target = country_data_processed_m7d['GDP_diff1'] # Mục tiêu là sai phân hiện tại
            X_m7d = country_data_processed_m7d[feature_columns_m7d]
            
            # 1e. Lấy giá trị GDP(t-1) tương ứng để tích hợp ngược
            # GDP_diff1(t) = GDP(t) - GDP(t-1)
            # => GDP(t) = GDP(t-1) + GDP_diff1(t)
            # Chúng ta cần GDP(t-1) tương ứng với mỗi y_m7d_diff_target
            # Vì country_data_processed_m7d đã dropna, index của nó có thể không liên tục với country_data_knn_m7d
            # Cách đơn giản là lấy GDP của dòng trước đó trong country_data_processed_m7d, nhưng nó sẽ là GDP(t-1) của chuỗi đã dropna
            # Cách chính xác hơn là tìm GDP(t-1) từ `country_data_knn_m7d` gốc dựa trên `Year`
            
            # Tạo một cột GDP_t_minus_1 trong country_data_processed_m7d
            # Nó sẽ là giá trị GDP của năm (Year - 1) từ country_data_knn_m7d gốc
            gdp_t_minus_1_list = []
            for current_year in country_data_processed_m7d['Year']:
                previous_year_gdp = country_data_knn_m7d[country_data_knn_m7d['Year'] == (current_year - 1)]['GDP']
                if not previous_year_gdp.empty:
                    gdp_t_minus_1_list.append(previous_year_gdp.iloc[0])
                else:
                    gdp_t_minus_1_list.append(np.nan) # Sẽ không xảy ra nếu dropna ở trên đúng
            country_data_processed_m7d['GDP_t_minus_1'] = gdp_t_minus_1_list
            
            # Lưu lại các giá trị này, chúng sẽ đi cùng với y_m7d_diff_target
            gdp_t_minus_1_for_integration_m7d = country_data_processed_m7d['GDP_t_minus_1']
            # Cũng lưu lại GDP thực tế tại thời điểm t (GDP_actual_t) để đánh giá cuối cùng
            gdp_actual_t_m7d = country_data_processed_m7d['GDP']


            print(f"\n--- Đặc trưng X (lags của sai phân) cho {MODEL_ID_M7D} (3 dòng đầu) ---"); display(X_m7d.head(3))
            print(f"\n--- Mục tiêu y (sai phân GDP_diff1) cho {MODEL_ID_M7D} (3 dòng đầu) ---"); display(y_m7d_diff_target.head(3).to_frame().T)
            print(f"\n--- GDP(t-1) để tích hợp ngược (3 dòng đầu) ---"); display(gdp_t_minus_1_for_integration_m7d.head(3).to_frame().T)
            print(f"\n--- GDP(t) thực tế (3 dòng đầu) ---"); display(gdp_actual_t_m7d.head(3).to_frame().T)
```

```python
  #CELL KNN_Diff2.M7D: {MODEL_ID_M7D} - CHIA TRAIN/TEST, CHUẨN HÓA VÀ HUẤN LUYỆN ---

model_knn_m7d = None 
X_train_m7d, X_test_m7d, y_train_m7d_diff, y_test_m7d_diff = None, None, None, None
gdp_t_minus_1_train_m7d, gdp_t_minus_1_test_m7d = None, None # Để tích hợp ngược
gdp_actual_t_train_m7d, gdp_actual_t_test_m7d = None, None # Để đánh giá cuối cùng
years_train_m7d, years_test_m7d = None, None
scaler_X_m7d = None 

if 'X_m7d' in locals() and X_m7d is not None and \
   'y_m7d_diff_target' in locals() and y_m7d_diff_target is not None and \
   'gdp_t_minus_1_for_integration_m7d' in locals() and gdp_t_minus_1_for_integration_m7d is not None and \
   'gdp_actual_t_m7d' in locals() and gdp_actual_t_m7d is not None and \
   not X_m7d.empty:
    
    split_index_m7d = int(len(X_m7d) * (1 - TEST_SIZE_RATIO_M7D))
    
    X_train_m7d, X_test_m7d = X_m7d.iloc[:split_index_m7d], X_m7d.iloc[split_index_m7d:]
    y_train_m7d_diff, y_test_m7d_diff = y_m7d_diff_target.iloc[:split_index_m7d], y_m7d_diff_target.iloc[split_index_m7d:]
    
    gdp_t_minus_1_train_m7d = gdp_t_minus_1_for_integration_m7d.iloc[:split_index_m7d]
    gdp_t_minus_1_test_m7d = gdp_t_minus_1_for_integration_m7d.iloc[split_index_m7d:]
    
    gdp_actual_t_train_m7d = gdp_actual_t_m7d.iloc[:split_index_m7d] # GDP thực tế tại t cho tập train
    gdp_actual_t_test_m7d = gdp_actual_t_m7d.iloc[split_index_m7d:]   # GDP thực tế tại t cho tập test

    years_full_processed_m7d = country_data_processed_m7d['Year']
    years_train_m7d = years_full_processed_m7d.iloc[:split_index_m7d]
    years_test_m7d = years_full_processed_m7d.iloc[split_index_m7d:]

    if X_train_m7d.empty or X_test_m7d.empty or len(X_test_m7d) < 1:
        print(f"Không đủ dữ liệu để chia train/test cho {MODEL_ID_M7D}.")
    else:
        print(f"{MODEL_ID_M7D} - Chia dữ liệu: Train {len(X_train_m7d)} mẫu, Test {len(X_test_m7d)} mẫu.")

        scaler_X_m7d = StandardScaler()
        X_train_scaled_m7d = scaler_X_m7d.fit_transform(X_train_m7d)
        X_test_scaled_m7d = scaler_X_m7d.transform(X_test_m7d)
        print("Đặc trưng X (lags của sai phân) đã được chuẩn hóa.")

        model_knn_m7d = KNeighborsRegressor(
            n_neighbors=N_NEIGHBORS_M7D,
            weights='distance', # Thử 'distance'
            metric='minkowski', p=2
        )
        
        print(f"\n--- {MODEL_ID_M7D}: Bắt đầu huấn luyện KNN Regressor (k={N_NEIGHBORS_M7D}) để dự đoán GDP_diff1 ---")
        model_knn_m7d.fit(X_train_scaled_m7d, y_train_m7d_diff) 
        print(f"Huấn luyện {MODEL_ID_M7D} hoàn tất.")
else:
    print(f"Biến đầu vào cho {MODEL_ID_M7D} chưa được tạo hoặc rỗng.")
```

```python
# --- CELL KNN_Diff3.M7D: {MODEL_ID_M7D} - ĐÁNH GIÁ MÔ HÌNH (SAU TÍCH HỢP NGƯỢC) ---

r2_knn_m7d, mae_knn_m7d, rmse_knn_m7d, mape_knn_m7d = np.nan, np.nan, np.nan, np.nan
y_pred_gdp_test_m7d = None; y_pred_gdp_train_m7d = None

if 'model_knn_m7d' in locals() and model_knn_m7d is not None and \
   'X_train_scaled_m7d' in locals() and X_train_scaled_m7d is not None and \
   'X_test_scaled_m7d' in locals() and X_test_scaled_m7d is not None and \
   'gdp_t_minus_1_train_m7d' in locals() and gdp_t_minus_1_train_m7d is not None and \
   'gdp_t_minus_1_test_m7d' in locals() and gdp_t_minus_1_test_m7d is not None and \
   'gdp_actual_t_train_m7d' in locals() and gdp_actual_t_train_m7d is not None and \
   'gdp_actual_t_test_m7d' in locals() and gdp_actual_t_test_m7d is not None and not gdp_actual_t_test_m7d.empty:
    
    # Dự đoán GDP_diff1
    y_pred_diff_train_m7d = model_knn_m7d.predict(X_train_scaled_m7d)
    y_pred_diff_test_m7d = model_knn_m7d.predict(X_test_scaled_m7d)

    # Tích hợp ngược để có dự đoán GDP
    # Predicted_GDP(t) = GDP(t-1)_actual + Predicted_GDP_diff1(t)
    y_pred_gdp_train_m7d = gdp_t_minus_1_train_m7d.values + y_pred_diff_train_m7d
    y_pred_gdp_test_m7d = gdp_t_minus_1_test_m7d.values + y_pred_diff_test_m7d
    
    # Chuyển thành Series để dễ xử lý index cho MAPE
    y_pred_gdp_train_m7d = pd.Series(y_pred_gdp_train_m7d, index=gdp_actual_t_train_m7d.index)
    y_pred_gdp_test_m7d = pd.Series(y_pred_gdp_test_m7d, index=gdp_actual_t_test_m7d.index)


    print(f"\n--- {MODEL_ID_M7D}: Đánh giá trên tập Huấn luyện (sau tích hợp ngược) ---")
    r2_train_m7d = r2_score(gdp_actual_t_train_m7d, y_pred_gdp_train_m7d)
    mae_train_m7d = mean_absolute_error(gdp_actual_t_train_m7d, y_pred_gdp_train_m7d)
    rmse_train_m7d = np.sqrt(mean_squared_error(gdp_actual_t_train_m7d, y_pred_gdp_train_m7d))
    try:
        mape_train_m7d = np.mean(np.abs((gdp_actual_t_train_m7d - y_pred_gdp_train_m7d) / (gdp_actual_t_train_m7d + 1e-9))) * 100
        print(f"R2: {r2_train_m7d:.4f}, MAE: {mae_train_m7d:.2f}, RMSE: {rmse_train_m7d:.2f}, MAPE: {mape_train_m7d:.2f}%")
    except Exception as e: print(f"R2: {r2_train_m7d:.4f}, MAE: {mae_train_m7d:.2f}, RMSE: {rmse_train_m7d:.2f}, Lỗi MAPE: {e}")

    print(f"\n--- {MODEL_ID_M7D}: Đánh giá trên tập Kiểm tra (sau tích hợp ngược) ---")
    r2_knn_m7d = r2_score(gdp_actual_t_test_m7d, y_pred_gdp_test_m7d)
    mae_knn_m7d = mean_absolute_error(gdp_actual_t_test_m7d, y_pred_gdp_test_m7d)
    rmse_knn_m7d = np.sqrt(mean_squared_error(gdp_actual_t_test_m7d, y_pred_gdp_test_m7d))
    
    y_true_m = gdp_actual_t_test_m7d.copy(); y_pred_m = y_pred_gdp_test_m7d.copy()
    zero_mask = (np.abs(y_true_m) < 1e-9)
    y_true_f = y_true_m[~zero_mask]; y_pred_f = y_pred_m[~zero_mask]
    if len(y_true_f) > 0: mape_knn_m7d = np.mean(np.abs((y_true_f - y_pred_f) / y_true_f)) * 100
    else: mape_knn_m7d = np.nan
            
    print(f"R-squared (Test): {r2_knn_m7d:.4f}")
    print(f"MAE (Test): {mae_knn_m7d:.2f}")
    print(f"RMSE (Test): {rmse_knn_m7d:.2f}")
    if not np.isnan(mape_knn_m7d): print(f"MAPE (Test): {mape_knn_m7d:.2f}% (Tính trên {len(y_true_f)} điểm)")
    else: print("MAPE (Test): Không thể tính.")

    plt.figure(figsize=(14, 7))
    if 'years_train_m7d' in locals() and years_train_m7d is not None:
        plt.plot(years_train_m7d, gdp_actual_t_train_m7d, label='GDP Huấn luyện (Thực tế)', color='royalblue', marker='o', markersize=4)
    if 'years_test_m7d' in locals() and years_test_m7d is not None:
        plt.plot(years_test_m7d, gdp_actual_t_test_m7d, label='GDP Kiểm tra (Thực tế)', color='forestgreen', marker='o', markersize=4)
        plt.plot(years_test_m7d, y_pred_gdp_test_m7d, label=f'GDP Kiểm tra (Dự đoán {MODEL_ID_M7D})', color='chocolate', marker='x', markersize=6, linestyle='--')
    plt.title(f'{MODEL_ID_M7D}: Đánh giá cho {COUNTRY_M7D} (k={N_NEIGHBORS_M7D}, Sai phân)', fontsize=15)
    plt.xlabel('Năm (Year)', fontsize=12); plt.ylabel('GDP (Tỷ Đô la Mỹ)', fontsize=12)
    plt.legend(fontsize=10); plt.grid(True, linestyle=':', alpha=0.7); plt.xticks(rotation=30, ha='right'); plt.tight_layout(); plt.show()
else:
    print(f"Mô hình {MODEL_ID_M7D} chưa được huấn luyện hoặc không có dữ liệu test/train để đánh giá.")
```

```python
# --- CELL KNN_Diff4.M7D: {MODEL_ID_M7D} - PHÂN TÍCH PHẦN DƯ ---
if 'model_knn_m7d' in locals() and model_knn_m7d is not None and \
   'y_train_m7d_diff' in locals() and y_train_m7d_diff is not None and 'y_pred_diff_train_m7d' in locals() and \
   'y_test_m7d_diff' in locals() and y_test_m7d_diff is not None and 'y_pred_diff_test_m7d' in locals() :
    
    print(f"\n\n--- {MODEL_ID_M7D}: PHÂN TÍCH PHẦN DƯ (CỦA GDP_diff1) CHO {COUNTRY_M7D.upper()} ---")
    residuals_diff_train_m7d = y_train_m7d_diff - y_pred_diff_train_m7d
    residuals_diff_test_m7d = y_test_m7d_diff - y_pred_diff_test_m7d

    # (Code vẽ các biểu đồ phần dư cho residuals_diff_train_m7d và residuals_diff_test_m7d)
    # Ví dụ: Biểu đồ Phần dư theo Thời gian
    plt.figure(figsize=(14, 6))
    if 'years_train_m7d' in locals() and years_train_m7d is not None:
        plt.plot(years_train_m7d, residuals_diff_train_m7d, marker='o', linestyle='-', color='blue', label='Phần dư GDP_diff1 Huấn luyện')
    if 'years_test_m7d' in locals() and years_test_m7d is not None and not residuals_diff_test_m7d.empty:
        plt.plot(years_test_m7d, residuals_diff_test_m7d, marker='x', linestyle='-', color='green', label='Phần dư GDP_diff1 Kiểm tra')
    plt.axhline(y=0, color='r', linestyle='--'); plt.xlabel('Năm'); plt.ylabel('Phần dư của GDP_diff1')
    plt.title(f'{MODEL_ID_M7D}: Phần dư của GDP_diff1 theo Thời gian cho {COUNTRY_M7D}'); plt.legend(); plt.grid(True); plt.tight_layout(); plt.show()
    # Thêm các biểu đồ khác: Residuals vs Predicted (của GDP_diff1), Histogram, Q-Q
else:
    print(f"Mô hình {MODEL_ID_M7D} chưa có đủ kết quả để phân tích phần dư.")
```
